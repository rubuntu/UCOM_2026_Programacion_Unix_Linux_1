El campo de la ciencia de datos está todavía en su infancia y, como tal, existen varias definiciones de lo que abarca.

## 1.4 ¿Por qué la ciencia de datos en la línea de comandos?

La línea de comandos tiene muchas ventajas que realmente pueden convertirlo en un proveedor de datos más eficiente y productivo.

científico. Agrupando aproximadamente las ventajas, la línea de comando es: ágil, aumentable, escalable, extensible y

omnipresente. A continuación explico cada ventaja.

### 1.4.1 La línea de comandos es ágil

La primera ventaja de la línea de comandos es que te permite ser ágil. La ciencia de datos tiene un enfoque muy interactivo y

La naturaleza exploratoria y el entorno en el que trabajas deben permitirlo. La línea de comando logra

esto por dos medios.

Primero, la línea de comando proporciona el llamado bucle de lectura-evaluación-impresión (REPL). Esto significa que escribe el comando,

presione Enter y el comando se evalúa inmediatamente. Un REPL suele ser mucho más conveniente para procesar datos.
ciencia que el ciclo editar-compilar-ejecutar-depurar asociado con scripts, programas grandes y, digamos, trabajos de Hadoop.

Sus comandos se ejecutan inmediatamente, se pueden detener a voluntad y se pueden cambiar rápidamente. este corto

El ciclo de iteración realmente te permite jugar con tus datos.

En segundo lugar, la línea de comando está muy cerca del sistema de archivos. Porque los datos son el ingrediente principal para hacer datos.

ciencia, es importante poder trabajar fácilmente con los archivos que contienen su conjunto de datos. La línea de comando ofrece

muchas herramientas convenientes para esto.

### 1.4.2 La línea de comando está aumentando

La línea de comando se integra bien con otras tecnologías. Cualquiera que sea la tecnología, su flujo de trabajo de ciencia de datos

actualmente incluye (ya sea R, Python o Excel), sepa que no estoy sugiriendo que abandone eso

flujo de trabajo. En su lugar, considere la línea de comando como una tecnología que amplifica las tecnologías

usted está empleando actualmente. Puede hacerlo de tres maneras.

En primer lugar, la línea de comandos puede actuar como pegamento entre muchas herramientas diferentes de ciencia de datos. Una forma de pegar herramientas es

conectar la salida de la primera herramienta a la entrada de la segunda herramienta. En el Capítulo 2 explico cómo funciona esto.

En segundo lugar, a menudo puedes delegar tareas a la línea de comandos desde tu propio entorno. Por ejemplo, Python, R,

y Apache Spark le permiten ejecutar herramientas de línea de comandos y capturar su salida. esto lo demuestro con

ejemplos en el Capítulo 10.

En tercer lugar, puede convertir su código (por ejemplo, un script de Python o R) en una herramienta de línea de comandos reutilizable. De esa manera,
Ya no importa en qué idioma esté escrito. Ahora, se puede utilizar directamente desde la línea de comandos o desde

cualquier entorno que se integre con la línea de comando como se menciona en el párrafo anterior. te explico como

esto en el Capítulo 4.

Al final, cada tecnología tiene sus fortalezas y debilidades, por lo que es bueno conocer varias y utilizar la que sea más adecuada.

más apropiado para la tarea en cuestión. A veces eso significa usar R, a veces la línea de comando, y

a veces incluso lápiz y papel. Al final de este libro, tendrá una comprensión sólida de cuándo debe usar

la línea de comando, y cuando sea mejor continuar con su lenguaje de programación favorito o

entorno informático.

### 1.4.3 La línea de comando es escalable

Como dije antes, trabajar en la línea de comandos es muy diferente a usar una GUI. En la línea de comando lo haces

cosas escribiendo, mientras que con una GUI, las cosas se hacen apuntando y haciendo clic con el mouse.

Todo lo que escriba manualmente en la línea de comando también se puede automatizar mediante scripts y herramientas. esto

hace que sea muy fácil volver a ejecutar sus comandos en caso de que haya cometido un error, cuando los datos de entrada hayan cambiado, o

porque su colega quiere realizar el mismo análisis. Además, sus comandos se pueden ejecutar en ubicaciones específicas.

intervalos, en un servidor remoto y en paralelo en muchos fragmentos de datos (más sobre esto en el Capítulo 8).

Debido a que la línea de comando es automatizable, se vuelve escalable y repetible. No es sencillo

automatizar apuntar y hacer clic, lo que hace que una GUI sea un entorno menos adecuado para realizar tareas escalables y repetibles.
ciencia de datos.

### 1.4.4 La línea de comando es extensible

La línea de comando en sí se inventó hace más de 50 años. Su funcionalidad principal se ha mantenido prácticamente sin cambios,

pero sus herramientas, que son los caballos de batalla de la línea de comandos, se desarrollan a diario.

La línea de comando en sí es independiente del idioma. Esto permite que las herramientas de línea de comandos se escriban en muchos idiomas diferentes.

lenguajes de programación. La comunidad de código abierto está produciendo muchos programas de línea de comandos gratuitos y de alta calidad.

herramientas que podemos utilizar para la ciencia de datos.

Estas herramientas de línea de comandos pueden funcionar juntas, lo que hace que la línea de comandos sea muy flexible. También puedes crear

sus propias herramientas, lo que le permite ampliar la funcionalidad efectiva de la línea de comando.

### 1.4.5 La línea de comando es omnipresente

Debido a que la línea de comando viene con cualquier sistema operativo tipo Unix, incluidos Ubuntu Linux y macOS, puede

encontrarse en muchos lugares. Además, el 100% de las 500 supercomputadoras principales ejecutan Linux.  Entonces, si alguna vez obtienes tu

9

manos en una de esas supercomputadoras (o si alguna vez te encuentras en Jurassic Park con las cerraduras de las puertas abiertas)

funcionando), ¡será mejor que conozcas la línea de comando!

Pero Linux no sólo se ejecuta en supercomputadoras. También se ejecuta en servidores, portátiles y sistemas integrados. Estos días,

Muchas empresas ofrecen computación en la nube, donde se pueden lanzar fácilmente nuevas máquinas sobre la marcha. Si alguna vez inicias sesión

a dicha máquina (o a un servidor en general), es casi seguro que llegará a la línea de comando.

También es importante tener en cuenta que la línea de comando no es sólo una exageración. Esta tecnología existe desde hace más
más de cinco décadas, y estoy convencido de que llegó para quedarse por otras cinco. Aprendiendo a usar la línea de comando

(para la ciencia de datos y en general) es, por tanto, una inversión que vale la pena.
