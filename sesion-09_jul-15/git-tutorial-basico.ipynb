| Tutorial |     | Básico |     |     | de Git |     |
| -------- | --- | ------ | --- | --- | ------ | --- |
Paso 1: Instalación
• Descarga e instala Git desde el sitio web oficial (https://git-scm.com/downloads) según tu sistema
operativo.
| Paso | 2: Configuración |             | inicial   |            |          |                        |
| ---- | ---------------- | ----------- | --------- | ---------- | -------- | ---------------------- |
|      | • Abre           | la terminal | (en       | Linux/Mac) | o el     | Git Bash (en Windows). |
|      | • Configura      |             | tu nombre | de         | usuario: |                        |


```bash
git config --global user.name "Tu nombre"
```

| Paso | 3: Crear     | un repositorio |            |             |                |                 |
| ---- | ------------ | -------------- | ---------- | ----------- | -------------- | --------------- |
|      | • Navega     | hasta          | la carpeta |             | de tu proyecto | en la terminal. |
|      | • Inicializa | un             | nuevo      | repositorio | Git:           |                 |


```bash
git init
```

| Paso | 4: Realizar | cambios    | y        | hacer  | commits            |     |
| ---- | ----------- | ---------- | -------- | ------ | ------------------ | --- |
|      | • Crea      | o modifica | archivos |        | en tu proyecto.    |     |
|      | • Verifica  | el         | estado   | actual | de tu repositorio: |     |


```bash
git status
```

Estos son solo los conceptos básicos para empezar a utilizar Git. Hay muchas otras funcionalidades y comandos
más avanzados que puedes explorar, como trabajar con ramas, resolver conflictos de fusión, deshacer cambios,
entre otros. Para obtener más información, te recomiendo consultar la documentación oficial de Git (https://git-
scm.com/doc).
