### Tutorial básico de Vim.

Vim, es un editor de texto altamente configurable y muy potente. Vim se basa en el editor de texto vi y está disponible en la mayoría de los sistemas Unix y Unix-like, así como en Windows.

1.  **Iniciar Vim**: Puedes iniciar Vim escribiendo "vim" seguido del nombre del archivo que deseas editar en la línea de comandos. Por ejemplo: `vim archivo.txt`. Si el archivo no existe, Vim lo creará automáticamente al guardarlo.

2.  **Modos de Vim**: Vim tiene dos modos principales: el modo de comandos y el modo de edición.

    -   **Modo de comandos**: Cuando inicias Vim, estás en el modo de comandos. Aquí puedes utilizar comandos para realizar acciones como guardar, salir o buscar texto.
    -   **Modo de edición**: Para comenzar a editar el archivo, debes cambiar al modo de edición. Presiona la tecla `i` para ingresar al modo de inserción y comenzar a escribir. Para volver al modo de comandos, presiona la tecla `Esc`.

3.  **Navegación básica**: Puedes moverte por el archivo utilizando las siguientes teclas:

    -   **h**: Mover el cursor hacia la izquierda.
    -   **j**: Mover el cursor hacia abajo.
    -   **k**: Mover el cursor hacia arriba.
    -   **l**: Mover el cursor hacia la derecha.

4.  **Guardar y salir**: Para guardar el archivo y salir de Vim, debes estar en el modo de comandos y ejecutar el comando `:wq` (escribe y sale). Si solo deseas guardar sin salir, utiliza el comando `:w`. Si deseas salir sin guardar, utiliza el comando `:q`. Si has realizado cambios y deseas salir sin guardar, utiliza el comando `:q!`.

5.  **Búsqueda de texto**: En el modo de comandos, puedes buscar texto utilizando el comando `/` seguido de la cadena de búsqueda. Por ejemplo, si deseas buscar la palabra "ejemplo", escribe `/ejemplo` y presiona Enter. Vim resaltará la primera ocurrencia encontrada. Para buscar la siguiente ocurrencia, presiona `n`.

6.  **Eliminar texto**: Puedes eliminar texto utilizando los siguientes comandos en el modo de comandos:

    -   **x**: Eliminar el carácter en la posición del cursor.
    -   **dd**: Eliminar la línea completa en la posición del cursor.

7.  **Copiar y pegar texto**: Puedes copiar y pegar texto utilizando los siguientes comandos en el modo de comandos:

    -   **yy**: Copiar la línea completa en la posición del cursor.
    -   **p**: Pegar el texto copiado después de la línea actual.

8.  **Deshacer y rehacer**: Para deshacer la última acción, presiona `u` en el modo de comandos. Para rehacer la última acción deshecha, presiona `Ctrl + r`.
