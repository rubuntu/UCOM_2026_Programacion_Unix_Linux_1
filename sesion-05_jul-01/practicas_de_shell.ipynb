## Pipes

Si no se usa PIPES se deben hacer varios pasos

```python
%%bash
ps > psout.txt
sort psout.txt > pssort.out

cat  pssort.out
```

```python
%%bash
ps | sort > pssort.out

cat  pssort.out
```

Para visualizar (sin usar cat, usando more)

```python
%%bash
ps | sort | more
```

Practicamnete no hay limites para conectar procesos

```python
%%bash
 ps  | sort | uniq | grep -v sh | more
```
