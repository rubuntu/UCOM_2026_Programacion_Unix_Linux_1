## Tutorial básico de Nano

Nano es un popular editor de texto en la línea de comandos para sistemas operativos basados en Unix. Es ligero, fácil de usar y viene preinstalado en muchas distribuciones de Linux. En este tutorial, te guiaré a través de los comandos básicos y la funcionalidad del editor Nano.

1. Abrir un archivo:
   Para abrir un archivo con Nano, simplemente escribe el siguiente comando en la terminal:
   ```
   nano nombre_archivo
   ```
   Reemplaza "nombre_archivo" por el nombre del archivo que deseas abrir o crea un archivo nuevo si no existe.

2. Navegación:
   Nano ofrece comandos sencillos de navegación para moverte por el texto:
   - Utiliza las teclas de flecha para mover el cursor hacia arriba, abajo, izquierda o derecha.
   - También puedes usar la combinación de teclas Ctrl+F para avanzar una página o Ctrl+B para retroceder una página.
   - Ctrl+V hace que el texto se desplace hacia abajo una página, y Ctrl+Y hacia arriba una página.

3. Edición de texto:
   Nano proporciona varios comandos de edición para modificar tu texto:
   - Simplemente comienza a escribir para insertar texto en la posición actual del cursor.
   - Usa la tecla Retroceso o Suprimir para eliminar caracteres.
   - El corte, copiado y pegado se realizan con Ctrl+K (cortar), Ctrl+U (copiar) y Ctrl+Shift+V (pegar), respectivamente.
   - Ctrl+_ (guión bajo) deshace la última acción.

4. Guardar y salir:
   Para guardar los cambios y salir de Nano, presiona Ctrl+O. Nano te solicitará el nombre del archivo si no lo especificaste anteriormente. Presiona Enter para confirmar el nombre del archivo.
   Para salir de Nano sin guardar, presiona Ctrl+X. Si tienes cambios no guardados, Nano te preguntará si deseas guardar antes de salir.

5. Buscar y reemplazar:
   Nano te permite buscar texto específico y reemplazarlo si es necesario:
   - Presiona Ctrl+W para activar la función de búsqueda. Ingresa el texto que deseas buscar y presiona Enter.
   - Para reemplazar el texto encontrado, presiona Ctrl+\ y ingresa el texto de reemplazo. Presiona Enter para reemplazar la primera ocurrencia o A para reemplazar todas las ocurrencias.

6. Configurar Nano:
   Nano se puede personalizar modificando el archivo de configuración llamado ".nanorc" en tu directorio de inicio. Este archivo te permite establecer opciones como resaltado de sintaxis, numeración de líneas y más.

Estos son los comandos y funcionalidades básicas del editor Nano. Nano también ofrece características adicionales que puedes explorar consultando la documentación o escribiendo "nano --help" en la terminal.
