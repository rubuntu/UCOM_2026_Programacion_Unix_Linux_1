### Pasar argumentos a programas

#### args.c: un programa que separa las opciones de la línea de comandos de los argumentos.

```
%%bash
cat args.c
```

```
%%bash
./args -i -lr ‘hi there’ -f fred.c
```

#### args.py : Version en Pyhton de args.c

```
%%bash
cat args.py
```

```
%%bash
./args.py Hola como estan!
```

#### argopt.c: una versión extendida de args.c que marca opciones ilegales y espera argumentos siguiendo algunas opciones.

```
%%bash
cat argopt.c
```

```
%%bash
./argopt -i -lr ‘hi there’ -f fred.c -q
```

#### argopt.py: Version en Python de argopt.c

```
%%bash
cat argopt.py
```

```
%%bash
./argopt.py -i entrada -o salida
```

#### longopt.c: una versión de argopt.c que usa getopt_long para admitir argumentos largos (entre comillas)

```
%%bash
cat longopt.c
```

```
%%bash
 ./longopt --init –l --file=fred.c 'hi there'
```

### Generación y configuración de mensajes de registro

#### syslog.c: un programa que escribe un mensaje en syslog en caso de error.

```
%%bash
cat syslog.c
```

```
%%bash
./syslog
sudo tail /var/log/messages 2>/dev/null | grep "oops - No such file or directory" >> output.txt
cat output.txt
```

#### logmask.c: un programa que usa logmask para registrar solo mensajes de cierta prioridad.

```
%%bash
cat logmask.c
```

```
%%bash
./logmask
# Ver logs informativos y de debug
sudo tail /var/log/messages 2>/dev/null | grep -e "[informative|debug] message" >> messages.txt
cat messages.txt
```

#### syslog.py - Version en Python de logmask.c

```
%%bash
cat syslog.py
```

```
%%bash
./syslog.py
# Ver logs informativos y de debug
sudo tail /var/log/syslog 2>/dev/null | grep -e "[Informative|Debug] message" >> messages.txt
cat messages.txt
```
