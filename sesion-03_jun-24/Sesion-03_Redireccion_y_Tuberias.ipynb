#### Descriptores estándares de archivo de entrada, salida y error

En los sistemas Unix y Unix-like, como Linux, los “standard I/O file descriptors” o “standard streams” son canales
de comunicación predefinidos que proporcionan una interfaz estándar para la entrada, salida y errores en un
programa. Estos canales son abiertos automáticamente por el sistema operativo cuando se ejecuta un programa
y están disponibles para su uso.

Hay tres “standard streams” predefinidos en Unix:

1. STDIN (Standard Input): Es el canal de entrada estándar que permite a un programa recibir datos de

entrada. Por defecto, el STDIN está conectado al teclado y el programa puede leer datos de entrada desde
allí. Sin embargo, STDIN también se puede redirigir para leer datos de un archivo o de otro programa a
través de tuberías (pipes).

2. STDOUT (Standard Output): Es el canal de salida estándar que permite a un programa enviar datos de
salida. Por defecto, el STDOUT está conectado a la pantalla y el programa puede imprimir datos en la
salida estándar, que se mostrarán en la consola. Al igual que STDIN, STDOUT también se puede redirigir
para enviar datos a un archivo o a otro programa a través de tuberías.

3. STDERR (Standard Error): Es el canal de salida de errores estándar que permite a un programa enviar
mensajes de error. Por defecto, el STDERR también está conectado a la pantalla y el programa puede
imprimir mensajes de error en el STDERR, que se mostrarán en la consola junto con la salida estándar. Al
igual que STDIN y STDOUT, STDERR también se puede redirigir para enviar mensajes de error a un
archivo o a otro programa a través de tuberías.

Estos tres canales (STDIN, STDOUT, STDERR) están asociados a los “file descriptors” en Unix:

STDIN tiene el “file descriptor” 0.
STDOUT tiene el “file descriptor” 1.
STDERR tiene el “file descriptor” 2.

Los programas en Unix interactúan con estos canales y “file descriptors” para leer datos de entrada, imprimir
datos de salida y mostrar mensajes de error. Pueden redirigir estos canales para leer o escribir en archivos en
lugar de la entrada y salida estándar, o para enviar datos a otros programas a través de tuberías.

La capacidad de redirigir y combinar estos canales es lo que permite la construcción de tuberías (pipes) y la
redirección de entrada/salida en los sistemas Unix. Esto brinda una gran flexibilidad y poder a la hora de
manipular el flujo de datos entre programas y archivos.

Los “standard I/O file descriptors” o “standard streams” en sistemas Unix son los canales de comunicación
predefinidos que proporcionan una interfaz estándar para la entrada (STDIN), salida (STDOUT) y errores
(STDERR) en un programa, permitiendo la lectura, escritura y redirección de datos.

### Redirección de entrada y salida

#### I. Concepto de redirección en sistemas Unix

En sistemas Unix, la redirección permite modificar la forma en que los comandos interactúan con los flujos
de entrada y salida estándar.
La redirección permite redirigir la salida de un comando a un archivo, o tomar la entrada de un archivo en
lugar del teclado.
Es una herramienta útil para el procesamiento de archivos y la automatización de tareas en la línea de
comandos.

#### II. Redirección de salida con el operador ‘>’

El operador ‘>’ se utiliza para redirigir la salida de un comando a un archivo.
Si el archivo existe, se sobrescribe con la nueva salida del comando.
Si el archivo no existe, se crea.
Ejemplo:  comando > archivo.txt

#### III. Redirección de salida con el operador ‘>>’

El operador ‘>>’ se utiliza para redirigir la salida de un comando y añadirla al final de un archivo existente.
Si el archivo no existe, se crea.
Ejemplo:  comando >> archivo.txt

#### IV. Redirección de entrada con el operador ‘<’

El operador ‘<’ se utiliza para tomar la entrada de un archivo en lugar del teclado.
El comando lee los datos del archivo en lugar de esperar la entrada del usuario.
Ejemplo:  comando < archivo.txt

#### V. Redirección de errores con el operador ‘2>’

El operador ‘2>’ se utiliza para redirigir los mensajes de error de un comando a un archivo.
Permite separar los errores de la salida estándar para facilitar su gestión.
Ejemplo:  comando 2> errores.txt

#### VI. Combinación de redirección de salida y errores con el operador ‘2>&1’

El operador ‘2>&1’ se utiliza para redirigir tanto la salida estándar como los mensajes de error a un mismo
archivo.
Permite agrupar la salida y los errores en un solo archivo para su análisis.
Ejemplo:  comando > salida_errores.txt 2>&1

### Tuberías (pipes)

#### I. Concepto de tuberías en sistemas Unix

En sistemas Unix, las tuberías (pipes) permiten redirigir la salida de un comando como entrada de otro
comando, creando una secuencia continua de operaciones.

Las tuberías permiten combinar de manera eficiente la salida de un comando con la entrada de otro, sin
necesidad de utilizar archivos temporales.

Una tubería Unix proporciona un flujo de datos unidireccional.

Una tubería se puede crear explícitamente en Unix usando la llamada al sistema de tuberías. Se devuelven dos
descriptores de archivo: fildes[0] y fildes[1], y ambos están abiertos para lectura y escritura. Una lectura de

lectura de fildes[1] accede a los datos escritos en fildes[0] también en FIFO base.

Cuando se utiliza una canalización en una línea de comandos de Unix, se supone que el primer proceso está
escribiendo en la salida estándar y que el segundo está leyendo desde la entrada estándar. Por lo tanto, es una
práctica común asignar el descriptor de dispositivo de escritura de tubería a stdout en el primer proceso y asignar
el descriptor de dispositivo de lectura de tubería a stdin en el segundo proceso.

## Ejemplos en Python

### I. Uso del módulo  os  para redirigir entrada y salida

El módulo  os  en Python proporciona funcionalidades para interactuar con el sistema operativo, incluyendo
la redirección de entrada y salida.
La función  os.dup2(fd, newfd)  permite redirigir un descriptor de archivo a otro.

### II. Implementación de tuberías en Python

En Python, se puede utilizar la funcionalidad de tuberías del sistema operativo para conectar la salida de un
proceso con la entrada de otro.
El módulo  subprocess  proporciona una forma conveniente de crear y gestionar procesos en Python.

### III. Ejemplos prácticos de redirección y tuberías en Python

#### Ejemplo 1: Redirigir la salida de un comando a un archivo en Python


```````python
#!/usr/bin/env python3
import os

# Abrir archivo de salida
file = open("archivo.txt", "w")

# Redirigir la salida estándar al archivo
os.dup2(file.fileno(), 1)

# Imprimir en la salida estándar (será redirigido al archivo)
print("Hola, mundo!")

# Cerrar archivo
file.close()
```````

```````python
! cat archivo.txt
```````

```````python
! echo "Sed ut perspiciatis unde omnis iste natus error sit voluptatem accusantium doloremque laudantium, totam rem aperiam, eaque ipsa quae ab illo inventore veritatis et quasi architecto beatae vitae dicta sunt explicabo. Nemo enim ipsam voluptatem quia voluptas sit aspernatur aut odit aut fugit, sed quia consequuntur magni dolores eos qui ratione voluptatem sequi nesciunt. Neque porro quisquam est, qui dolorem ipsum quia dolor sit amet, consectetur, adipisci velit, sed quia non numquam eius modi tempora incidunt ut labore et dolore magnam aliquam quaerat voluptatem. Ut enim ad minima veniam, quis nostrum exercitationem ullam corporis suscipit laboriosam, nisi ut aliquid ex ea commodi consequatur?  " >>  lorem-ipsum.txt
```````

```````python
#!/usr/bin/env python3
import subprocess

# Crear proceso 1 para leer el archivo
proceso1 = subprocess.Popen(["cat", "lorem-ipsum.txt"], stdout=subprocess.PIPE)

# Crear proceso 2 para contar las líneas
proceso2 = subprocess.Popen(["wc", "-l"], stdin=proceso1.stdout, stdout=subprocess.PIPE)

# Leer la salida del proceso 2
salida = proceso2.communicate()[0]

# Imprimir el número de líneas
print("Número de líneas:", salida.decode().strip())
```````

```
#!/usr/bin/env python3
import sys

def imprimir_contenido(archivo):
    with open(archivo, 'r') as file:
        contenido = file.read()
        print(contenido)

# Verificar si se proporcionó un archivo como argumento
if len(sys.argv) > 1:
    archivo = sys.argv[1]
    imprimir_contenido(archivo)
else:
    # Leer desde stdin si no se proporciona un archivo como argumento
    contenido = sys.stdin.read()
    print(contenido)
```

En este programa, se utiliza la función  imprimir_contenido  para leer y mostrar el contenido de un archivo. Si se
proporciona un archivo como argumento al ejecutar el programa, se lee ese archivo y se imprime su contenido. Si
no se proporciona ningún archivo como argumento, el programa lee desde la entrada estándar (stdin) y muestra
el contenido.

b. Implementa una función en Python que reciba una cadena por la entrada estándar (stdin) o como primer

argumento y devuelva la cantidad de palabras que contiene


```
#!/usr/bin/env python3
import sys

def contar_palabras(cadena):
    # Eliminar los espacios en blanco al inicio y final de la cadena
    cadena = cadena.strip()

    # Dividir la cadena en palabras utilizando los espacios en blanco como separadores
    palabras = cadena.split()

    # Devolver la cantidad de palabras
    return len(palabras)

# Verificar si se proporcionó una cadena como argumento
if len(sys.argv) > 1:
    cadena = ' '.join(sys.argv[1:])
else:
    # Leer desde stdin si no se proporciona una cadena como argumento
    cadena = sys.stdin.read()

cantidad_palabras = contar_palabras(cadena)
print("Cantidad de palabras:", cantidad_palabras)
```

En este programa, la función  contar_palabras  recibe una cadena y realiza los siguientes pasos: 1. Elimina los
espacios en blanco al inicio y al final de la cadena utilizando el método  strip() . 2. Divide la cadena en palabras
utilizando el método  split() , que utiliza los espacios en blanco como separadores. 3. Devuelve la cantidad de
palabras obtenidas utilizando la función  len() .

Luego, el programa verifica si se proporcionó una cadena como argumento al ejecutarlo. Si se proporciona, se
utiliza esa cadena. Si no se proporciona, se lee desde la entrada estándar (stdin). Luego, se llama a la función
contar_palabras  con la cadena y se imprime la cantidad de palabras obtenida.
