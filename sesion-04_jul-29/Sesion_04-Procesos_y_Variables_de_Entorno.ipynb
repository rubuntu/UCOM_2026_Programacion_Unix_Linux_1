### 1. Introducción a los procesos en Unix/Linux

-   Concepto de proceso y su importancia en el sistema operativo    
-   Jerarquía de procesos: proceso padre y proceso hijo    
-   Identificación de procesos: PID (Identificador de proceso)    
-   Comandos básicos: ps, top, pstree    

Concepto de proceso y su importancia en el sistema operativo: - Un proceso en Unix/Linux es una instancia de un programa en ejecución. Cada vez que ejecutas un programa, se crea un proceso para ese programa.
- Los procesos son fundamentales en el sistema operativo, ya que permiten la ejecución simultánea de múltiples programas y la gestión eficiente de los recursos del sistema.     
- Cada proceso tiene su propio espacio de memoria y recursos asignados, y se ejecuta de forma independiente de otros procesos en el sistema.    

Jerarquía de procesos: proceso padre y proceso hijo:      
- En Unix/Linux, los procesos forman una jerarquía en forma de árbol, donde cada proceso tiene un proceso padre, excepto el proceso raíz, que es el primer proceso lanzado durante el inicio del sistema.
- Cuando se crea un nuevo proceso, este puede generar procesos hijos. Estos procesos hijos heredan ciertos atributos y recursos del proceso padre.     
- Los procesos hijos pueden a su vez generar más procesos hijos, lo que da lugar a una jerarquía de procesos.    

Identificación de procesos: PID (Identificador de proceso):
- Cada proceso en Unix/Linux se identifica mediante un número único llamado PID (Process ID).     
- El PID permite realizar operaciones específicas en un proceso, como enviar señales para terminar su ejecución o obtener información sobre el proceso.     
- Puedes obtener el PID de un proceso utilizando el comando `ps` o a través de funciones de programación específicas.     

Comandos básicos para trabajar con procesos:

1.  Comando `ps` (Process Status):

    -   Muestra información sobre los procesos en ejecución.
    -   Ejemplo: `ps aux` muestra una lista de todos los procesos del sistema junto con detalles como el PID, el consumo de CPU, la memoria utilizada, etc.

2.  Comando `top`:

    -   Proporciona una visión dinámica en tiempo real de los procesos en ejecución.
    -   Muestra información en forma de tabla, con actualizaciones periódicas.
    -   Proporciona detalles sobre el uso de CPU, memoria y otros recursos del sistema.

3.  Comando `pstree`:

    -   Muestra una representación en forma de árbol de la jerarquía de procesos en el sistema.
    -   Puedes visualizar la relación entre los procesos padres e hijos.

A continuación, te muestro ejemplos de cómo usar estos comandos en la línea de comandos:

-   Ejemplo de uso del comando `ps`:

````python
%%bash
  ps | head -n 10
````

-   Ejemplo de uso del comando `top`:

````python
%%bash
# top
````

-   Ejemplo de uso del comando `pstree`:

````python
%%bash
  pstree
````

Estos comandos te brindan una visión general de los procesos en ejecución y su relación, pero hay muchas opciones y combinaciones posibles que se puede explorar consultando la documentación de cada comando.

### 2. Variables de entorno

-   Definición y uso de variables de entorno en Unix/Linux    
-   Acceso a variables de entorno desde la línea de comandos    
-   Variables de entorno predefinidas: PATH, HOME, USER, etc.    
-   Establecimiento de variables de entorno personalizadas    

Durante esta parte de la sesión, nos enfocaremos en las variables de entorno en Unix/Linux. Exploraremos su definición y uso, cómo acceder a ellas desde la línea de comandos, algunas variables de entorno predefinidas comunes, y cómo establecer nuestras propias variables de entorno personalizadas.

Definición y uso de variables de entorno en Unix/Linux:     
- Las variables de entorno son valores dinámicos que pueden afectar el comportamiento de los programas en un sistema Unix/Linux.     
- Estas variables se utilizan para almacenar información importante como rutas de búsqueda, configuraciones, preferencias del usuario, entre otros.
- Los programas pueden acceder y utilizar estas variables para adaptar su comportamiento o acceder a información específica.

Acceso a variables de entorno desde la línea de comandos:     
- Para acceder al valor de una variable de entorno en la línea de comandos, se utiliza el símbolo de dólar `$` seguido del nombre de la variable.     
- Ejemplo: `echo $PATH` muestra el valor de la variable de entorno PATH, que almacena las rutas de búsqueda de comandos.

Variables de entorno predefinidas:      
- Unix/Linux tiene varias variables de entorno predefinidas que proporcionan información útil sobre el sistema y el usuario actual.     
- Algunas variables de entorno comunes son:
  - PATH: Almacena las rutas de búsqueda de comandos.
  - HOME: Ruta al directorio de inicio del usuario actual.
  - USER: Nombre de usuario del usuario actual.
  - SHELL: Ruta al shell (intérprete de comandos) predeterminado del usuario actual.

Establecimiento de variables de entorno personalizadas:
- Puedes establecer tus propias variables de entorno personalizadas para almacenar valores específicos que deseas utilizar en programas o scripts.
- En la línea de comandos, puedes establecer una variable de entorno usando el formato `nombre_variable=valor`.
- Ejemplo: `export MY_VAR="Hello, World!"` establece una variable de entorno llamada MY_VAR con el valor "Hello, World!".

A continuación, te muestro ejemplos de cómo acceder a variables de entorno predefinidas y cómo establecer variables de entorno personalizadas desde la línea de comandos:

-   Ejemplo de acceso a variables de entorno predefinidas:

````python
%%bash
echo $PATH

echo $HOME

echo $USER
````

-   Ejemplo de establecimiento de una variable de entorno personalizada:

````python
%%bash
export MY_VAR="Hello, World!"

echo $MY_VAR
````

Si deseas utilizar variables de entorno en programas o scripts escritos en un lenguaje de programación como Python, puedes acceder a ellas a través de las funciones y bibliotecas específicas proporcionadas por ese lenguaje. A continuación, te muestro un ejemplo de cómo acceder a una variable de entorno en Python:

````python
%%python
import os

# Acceso a una variable de entorno
path = os.environ.get('PATH')
print(path)

# Establecimiento de una variable de entorno personalizada
os.environ['MY_VAR'] = 'Hello, World!'
````

Las variables de entorno pueden variar entre diferentes sistemas operativos y configuraciones, por lo que es importante consultar la documentación específica de tu sistema para obtener más detalles y opciones relacionadas con las variables de entorno.

### 3. Comandos básicos relacionados con procesos y variables de entorno

-   Comando ps: visualización de procesos en ejecución
-   Comando kill: terminación de procesos
-   Comando env: visualización de variables de entorno
-   Comando export: establecimiento de variables de entorno

1.  Comando `ps` (Process Status):

    -   El comando `ps` se utiliza para mostrar información sobre los procesos en ejecución en el sistema.
    -   Puedes utilizar diferentes opciones con `ps` para personalizar la información que deseas mostrar.
    -   Ejemplo: `ps aux` muestra una lista de todos los procesos del sistema con detalles como el PID, el consumo de CPU, la memoria utilizada, etc.

    Ejemplo de uso del comando `ps`:


````python
%%bash
ps aux | head -n 10
````

````python
%%bash
ps aux | grep watch
````

2.  Comando `kill`:

    -   El comando `kill` se utiliza para terminar un proceso específico en el sistema.
    -   Se le pasa como argumento el PID del proceso que se desea terminar.
    -   El PID se puede obtener mediante el comando `ps` u otras herramientas de monitoreo de procesos.
    -   Ejemplo: `kill 1234` termina el proceso con el PID 1234.

    Ejemplo de uso del comando `kill`:


````python
%%bash
kill 28952
````

3.  Comando `env`:

    -   El comando `env` se utiliza para mostrar las variables de entorno establecidas en el sistema.
    -   Muestra una lista de variables de entorno con sus valores.
    -   Ejemplo: `env` muestra todas las variables de entorno disponibles.

    Ejemplo de uso del comando `env`:


````python
%%bash
env | head -n 10
````

4.  Comando `export`:

    -   El comando `export` se utiliza para establecer variables de entorno personalizadas.
    -   Permite definir una variable de entorno y asignarle un valor.
    -   La variable de entorno establecida con `export` estará disponible para todos los procesos lanzados desde la sesión actual.
    -   Ejemplo: `export MY_VAR="Hello, World!"` establece una variable de entorno llamada MY_VAR con el valor "Hello, World!".

    Ejemplo de uso del comando `export`:


````python
%%bash
export MY_VAR="Hello, World!"
````

````python
%%bash
echo $MY_VAR
````

Otro ejemplo de uso del comando `export`:

````python
%%bash
#!/bin/bash

export var1="Esta variable es global"
var2="Variable2"

principal() {

    bash -c '
    echo -e "\nEn la parte principal"
    echo "Imprimiendo \$var1: $var1"
    echo "Imprimiendo \$var2: $var2"'
}

principal
export var2  # A partir de aquí ya debería aparecer el valor de "var2"
principal
````

### 4. Ejemplos en Python

-   Uso del módulo `os` en Python para trabajar con procesos y variables de entorno
-   Obtención de información sobre el proceso actual
-   Creación de procesos hijos desde Python
-   Manipulación de variables de entorno desde Python

1.  Obtención de información sobre el proceso actual:

    -   Utilizando el módulo `os`, puedes obtener información sobre el proceso actual, como el ID del proceso (PID) y el directorio de trabajo actual.
    -   El PID del proceso actual se puede obtener utilizando la función `os.getpid()`.
    -   El directorio de trabajo actual se puede obtener utilizando la función `os.getcwd()`.

    Ejemplo de obtención de información sobre el proceso actual en Python:


````python
%%python
import os

# Obtención del PID del proceso actual
pid = os.getpid()
print("PID:", pid)

# Obtención del directorio de trabajo actual
cwd = os.getcwd()
print("Directorio actual:", cwd)
````

2.  Creación de procesos hijos desde Python:

    -   Utilizando el módulo `os`, puedes crear procesos hijos desde Python.
    -   La función `os.fork()` se utiliza para crear un nuevo proceso hijo.
    -   Después de la llamada a `fork()`, el proceso se divide en dos, y tanto el proceso padre como el proceso hijo continúan la ejecución a partir de ese punto.
    -   La función `os.fork()` devuelve 0 en el proceso hijo y el PID del hijo en el proceso padre.

    Ejemplo de creación de un proceso hijo en Python:

    ```{python}

    import os
    pid = os.fork()
    if pid == 0:  
      print("Proceso hijo")  
    else:
      print("Proceso padre")
        
    ```

3.  Manipulación de variables de entorno desde Python:

    -   Utilizando el módulo `os`, puedes manipular las variables de entorno desde Python.
    -   La función `os.environ` proporciona un diccionario que contiene todas las variables de entorno actuales.
    -   Puedes acceder y modificar las variables de entorno utilizando este diccionario.

    Ejemplo de manipulación de variables de entorno en Python:


````python
%%python
import os

# Obtención del valor de una variable de entorno
path = os.environ.get('PATH')
print("PATH:", path)

# Establecimiento de una variable de entorno personalizada
os.environ['MY_VAR'] = 'Hello, World!'
hello=os.environ.get('MY_VAR')
print("MY_VAR:", hello)


````
