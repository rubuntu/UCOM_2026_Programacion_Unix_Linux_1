### Manipulación de archivos y directorios en Unix/Linux


#### Comandos básicos en bash shell:

##### 1.  ls: Listar archivos y directorios.


```python
%%bash
ls
```

Este comando lista los archivos y directorios en el directorio actual.

##### 2.  cd: Cambiar de directorio.


```python
%%bash
cd /content/sample_data
```

Este comando cambia el directorio actual al especificado.

##### 3.  pwd: Mostrar el directorio actual.


```python
%%bash
pwd
```

Este comando muestra el directorio actual en el que te encuentras.

##### 4.  mkdir: Crear directorios.


```python
%%bash
mkdir directorio
ls 
```

Este comando crea un nuevo directorio con el nombre especificado.

##### 5.  rm: Eliminar archivos y directorios.


```python
%%bash
touch archivo
ls
```

```python
%%bash
rm archivo
ls
```

Este comando elimina el archivo especificado.


```python
%%bash
rm -r directorio
ls 
```

Este comando elimina el directorio y su contenido de forma recursiva.

##### 6.  cp: Copiar archivos y directorios.


```python
%%bash
touch archivo
cp archivo archivo-destino
ls 
```

Este comando copia el archivo al destino especificado.


```python
%%bash
mkdir directorio
cp -r directorio directorio-destino
ls 
```

Este comando copia el directorio y su contenido de forma recursiva al destino especificado.

##### 7.  mv: Mover o renombrar archivos y directorios.


```python
%%bash
mv archivo archivo-renombrado
ls 
```

Este comando mueve o renombra el archivo al destino especificado.


```python
%%bash
mv directorio directorio-renombrado
ls 
```

Este comando mueve o renombra el directorio al destino especificado.

Tener en cuenta que estos son solo ejemplos básicos de los comandos en{bash} shell. Cada comando tiene opciones adicionales que puedes explorar utilizando el manual de Unix/Linux mediante el comando `man` seguido del nombre del comando, por ejemplo, `man ls` o `man cp`.


### Permisos y propietarios de archivos

**Concepto de permisos en Unix/Linux: lectura, escritura y ejecución:** En Unix/Linux, cada archivo y directorio tiene asociados permisos que determinan qué acciones pueden realizarse sobre ellos. Los permisos se dividen en tres categorías:

1.  Lectura (Read): Permite leer el contenido del archivo o ver los nombres de los archivos en un directorio.
2.  Escritura (Write): Permite modificar el contenido de un archivo o crear, eliminar y renombrar archivos en un directorio.
3.  Ejecución (Execute): Permite ejecutar un archivo como un programa o ingresar a un directorio.

Los permisos se representan con letras o símbolos especiales y se aplican a tres entidades principales: el propietario del archivo, el grupo asociado al archivo y otros usuarios.

B. Permisos para el propietario, grupo y otros usuarios: Cada entidad tiene permisos específicos asignados, y estos permisos se representan utilizando símbolos y se combinan para formar una cadena de permisos.

-   Propietario (Owner):
    -   Lectura: r
    -   Escritura: w
    -   Ejecución: x
-   Grupo (Group):
    -   Lectura: r
    -   Escritura: w
    -   Ejecución: x
-   Otros (Others):
    -   Lectura: r
    -   Escritura: w
    -   Ejecución: x

Por ejemplo, si un archivo tiene permisos rw-r--r--, significa que el propietario tiene permisos de lectura y escritura, el grupo tiene permiso de lectura solamente, y otros usuarios también tienen permiso de lectura solamente.

C. Comando chmod: Cambiar los permisos de un archivo o directorio: El comando `chmod` se utiliza para cambiar los permisos de un archivo o directorio en Unix/Linux. Los permisos pueden establecerse utilizando números octales o símbolos especiales.

1.  Establecer permisos utilizando números octales:


```python
%%bash
touch archivo.txt
chmod 755 archivo.txt
ls -l archivo.txt
```

Este comando establece los permisos en rw-r-xr-x para el archivo.txt. El número 7 representa los permisos del propietario, el número 5 representa los permisos del grupo y otros usuarios.

2.  Establecer permisos utilizando símbolos especiales:


```python
%%bash
chmod u+rwx,go-wx archivo.txt
ls -l archivo.txt
```

Este comando establece permisos de lectura, escritura y ejecución para el propietario y permisos de solo lectura para el grupo y otros usuarios.

D. Comando chown: Cambiar el propietario y el grupo de un archivo o directorio: El comando `chown` se utiliza para cambiar el propietario y el grupo de un archivo o directorio en Unix/Linux.

*Observación:* Por seguridad en muchos sistemas \*nix el comando **chown** solo puede ejecutarse por el usuario "root" o por un usuario "sudoer"

1.  Cambiar el propietario:


```python
%%bash
touch archivo2.txt
sudo chown postgres archivo2.txt
ls -l  archivo2.txt
```

Este comando cambia el propietario del archivo2.txt al nuevo_propietario "postgres".

2.  Cambiar el grupo:


```python
%%bash
sudo chown :postgres archivo2.txt
ls -l  archivo2.txt
```

Este comando cambia el grupo del archivo.txt al nuevo_grupo "postgres".

3.  Cambiar el propietario y el grupo:


```python
%%bash
sudo chown ruben:ruben archivo2.txt
ls -l  archivo2.txt
```

Este comando cambia tanto el propietario como el grupo del archivo2.txt.

Recuerdar que los permisos y propietarios son elementos cruciales en Unix/Linux para garantizar la seguridad y el acceso adecuado a los archivos y directorios. A través de los comandos `chmod` y `chown`, puedes modificar y ajustar estos aspectos según tus necesidades.


### Ejemplos en Python


#### Breve introducción a la manipulación de archivos en Python.

-   En Python, la manipulación de archivos se realiza utilizando funciones y métodos proporcionados por la biblioteca estándar.

-   Antes de interactuar con un archivo, es necesario abrirlo en el modo adecuado (lectura, escritura, etc.).

-   Es importante cerrar los archivos una vez finalizada la manipulación para liberar los recursos del sistema.


#### Ejemplos prácticos de lectura, escritura y manipulación de archivos en Python.

-   Ejemplo: Lectura de un archivo


```python
%%bash
echo "Hola Mundo" > archivo.txt
```

```python
# Abrir el archivo en modo lectura
archivo = open("archivo.txt", "r")

# Leer todo el contenido del archivo
contenido = archivo.read()

# Imprimir el contenido del archivo
print(contenido)

# Cerrar el archivo
archivo.close()
```

2.  Ejemplo: Escritura en un archivo


```python
# Abrir el archivo en modo escritura
archivo = open("archivo.txt", "w")

# Escribir contenido en el archivo
archivo.write("Hola, este es un ejemplo de escritura en un archivo.")

# Cerrar el archivo
archivo.close()
```

3.  Ejemplo: Manipulación de archivos utilizando el contexto de "with"


```python
# Abrir el archivo en modo lectura utilizando el contexto "with"
with open("archivo.txt", "r") as archivo:
    # Leer todo el contenido del archivo
    contenido = archivo.read()

# Imprimir el contenido del archivo
print(contenido)
# El archivo se cierra automáticamente al salir del bloque "with"
```

C. Ejemplo de uso de la biblioteca os para manipular archivos y directorios en Unix/Linux:


```python
import os

# Obtener el directorio actual
directorio_actual = os.getcwd()
print("Directorio actual:", directorio_actual)

# Crear un nuevo directorio
nuevo_directorio = "nuevo_directorio"
os.mkdir(nuevo_directorio)
print("Directorio creado:", nuevo_directorio)

# Mover un archivo a otro directorio
archivo_origen = "archivo.txt"
directorio_destino = "nuevo_directorio"
os.rename(archivo_origen, os.path.join(directorio_destino, archivo_origen))
print("Archivo movido al directorio:", directorio_destino)

# Eliminar un archivo
archivo_eliminar = os.path.join(directorio_destino, archivo_origen)
os.remove(archivo_eliminar)
print("Archivo eliminado:", archivo_eliminar)

# Eliminar un directorio
os.rmdir(nuevo_directorio)
print("Directorio eliminado:", nuevo_directorio)
```

En este ejemplo, se utilizan las funciones de la biblioteca `os` para obtener el directorio actual, crear un nuevo directorio, mover un archivo a otro directorio, eliminar un archivo y eliminar un directorio.

Recuerdar que es importante tener los permisos adecuados para realizar estas operaciones en el sistema de archivos.
