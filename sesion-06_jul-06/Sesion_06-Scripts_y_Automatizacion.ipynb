### 1. Introducción a los scripts en Unix/Linux

* Definición de un script y su importancia en la automatización de
  tareas.
* Ventajas de utilizar scripts en entornos Unix/Linux.
* Tipos de scripts más comunes en Unix/Linux (bash, sh, csh,
  etc.).
* Menciones de casos de uso comunes para scripts en Unix/Linux.

---

* Definición de un script y su importancia en la automatización de
  tareas: Un script en Unix/Linux es un archivo de texto que contiene una
  secuencia de comandos y/o instrucciones que se ejecutan de manera
  secuencial. Estos scripts permiten automatizar tareas repetitivas o
  complejas, facilitando la administración y configuración del
  sistema.
* Ventajas de utilizar scripts en entornos Unix/Linux:

  + Ahorro de tiempo y esfuerzo: Los scripts permiten realizar tareas
    complejas en forma automatizada, lo que reduce la necesidad de ejecutar
    comandos manualmente.
  + Reproducibilidad: Los scripts pueden ser compartidos y ejecutados en
    diferentes sistemas, asegurando que las tareas se realicen de manera
    consistente.
  + Facilidad de mantenimiento: Los scripts pueden ser modificados y
    mejorados con facilidad, lo que facilita la actualización y adaptación a
    cambios en los requerimientos.
  + Automatización de tareas repetitivas: Los scripts son ideales para
    automatizar tareas que se deben realizar de manera frecuente, evitando
    errores humanos y mejorando la eficiencia.
* Tipos de scripts más comunes en Unix/Linux: Los scripts en
  Unix/Linux pueden ser escritos en diferentes lenguajes de scripting,
  entre los cuales los más comunes son:

  + Bash (Bourne Again Shell): Es el shell por defecto en la mayoría de
    las distribuciones de Unix/Linux. Ofrece una amplia gama de
    características y es ampliamente utilizado.
  + Sh (Bourne Shell): Es un shell más básico y simple en comparación
    con Bash. Aunque menos común hoy en día, aún se utiliza en algunos
    sistemas.
  + Csh (C Shell): Es un shell con una sintaxis similar a C. Aunque no
    es tan popular como Bash, sigue siendo utilizado en algunos sistemas
    Unix.
  + Otros lenguajes de scripting como Python, Perl, Ruby, entre otros,
    también se utilizan para escribir scripts en Unix/Linux.
* Menciones de casos de uso comunes para scripts en Unix/Linux:

  + Tareas de administración del sistema: Los scripts pueden automatizar
    la instalación de software, configuración de servicios, respaldos,
    gestión de usuarios, entre otras tareas relacionadas con la
    administración del sistema.
  + Procesamiento y manipulación de archivos: Los scripts permiten
    realizar operaciones como búsqueda, filtrado, transformación y
    generación de informes sobre archivos y datos.
  + Monitoreo y generación de informes: Los scripts pueden ser
    utilizados para recolectar y procesar datos de monitoreo del sistema,
    generando informes y alertas automatizadas.
  + Despliegue de aplicaciones: Los scripts pueden facilitar el
    despliegue y configuración automatizada de aplicaciones en
    servidores.
  + Automatización de pruebas: Los scripts pueden automatizar la
    ejecución de pruebas de software, agilizando el proceso de control de
    calidad.

A continuación, se muestra un ejemplo básico de un script en Bash que
imprime “Hola, mundo!”:




````bash
%%sh

#!/bin/bash
echo "Hola, mundo!"

echo 'echo "Hola, mundo!" ' > hola.sh
````



Este script se puede guardar en un archivo con extensión “.sh” (por
ejemplo, “hola.sh”), y luego ejecutarlo en una terminal con el siguiente
comando:




````bash
%%sh

bash hola.sh
````



Al ejecutar el script, se mostrará el mensaje “Hola, mundo!” en la
terminal.

### 2. Shell scripting: sintaxis básica, variables, bucles, condicionales (60 minutos)

* Sintaxis básica de un script en shell.
* Declaración e inicialización de variables en shell scripting.
* Uso de bucles (for, while) y condicionales (if, case) en shell
  scripting.
* Ejemplos prácticos de shell scripting para realizar tareas

---

* Sintaxis básica de un script en shell: Un script en shell se
  inicia con una línea de shebang (#!/bin/bash) que indica el intérprete a
  utilizar. A continuación, se escriben los comandos y las instrucciones
  del script.
* Declaración e inicialización de variables en shell scripting: En
  shell scripting, las variables se declaran y se asignan valores
  utilizando el formato `nombre_variable=valor`. No se debe
  dejar espacios alrededor del signo igual (=). Para acceder al valor de
  una variable, se utiliza el signo de dólar ($) seguido del nombre de la
  variable. Por ejemplo:




````bash
%%bash
nombre="Juan"
edad=25
echo "Mi nombre es $nombre y tengo $edad años."
````



* Uso de bucles (for, while) y condicionales (if, case) en shell
  scripting:
  + Bucles for: Permiten repetir una serie de comandos para un conjunto
    de valores. Por ejemplo:




````bash
%%bash
for i in 1 2 3 4 5
do
   echo "El valor es: $i"
done

````



* Bucles while: Repiten una serie de comandos mientras se cumpla una
  condición. Por ejemplo:




````bash
#%% shell
contador=0
while [ $contador -lt 5 ]
do
   echo "El contador es: $contador"
   contador=$((contador+1))
done

````



* Condicionales if: Permite ejecutar diferentes comandos según una
  condición. Por ejemplo:




````bash
%%bash
edad=18
if [ $edad -ge 18 ]
then
   echo "Eres mayor de edad."
else
   echo "Eres menor de edad."
fi

````



* Condicionales case: Permite ejecutar diferentes comandos según el
  valor de una variable. Por ejemplo:




````bash
%%bash

opcion="A"
case $opcion in
   A)
      echo "Seleccionaste la opción A."
      ;;
   B)
      echo "Seleccionaste la opción B."
      ;;
   *)
      echo "Opción no válida."
      ;;
esac

````



* Ejemplos prácticos de shell scripting para realizar tareas
  sencillas: Ejemplo 1: Script que muestra los números pares del 1 al
  10:




````bash
%%bash

#!/bin/bash
for (( i=1; i<=10; i++ ))
do
   if (( $i % 2 == 0 ))
   then
      echo $i
   fi
done

````

2. Ejemplo de script para procesamiento de archivos:




````bash
%%bash

# 1. Definir primero las variables de ruta
directorio_entrada="./directorio_entrada"
directorio_salida="./directorio_salida"

# 2. Asegurar que los directorios existan (el flag -p evita errores si ya existen)
mkdir -p "$directorio_entrada"
mkdir -p "$directorio_salida"

# 3. Crear el archivo de prueba DENTRO del directorio ya creado
echo "hola mundo, esto es una prueba" > "$directorio_entrada/archivo1.csv"

# 4. Procesar los archivos
for archivo in "$directorio_entrada"/*
do
    # Validar que realmente sea un archivo (por si el directorio está vacío)
    [ -e "$archivo" ] || continue

    # Obtener solo el nombre del archivo sin la ruta
    nombre_archivo=$(basename "$archivo")

    # Procesar (convertir a mayúsculas)
    tr '[:lower:]' '[:upper:]' < "$archivo" > "$directorio_salida/$nombre_archivo"
done

# Verificación de que funcionó:
echo "--- Contenido en salida ---"
cat "$directorio_salida/archivo1.csv"
````



Este script procesa cada archivo en un directorio de entrada
especificado y realiza una acción deseada, como convertir el contenido a
mayúsculas. Los archivos procesados se guardan en un directorio de
salida especificado.

Recuerda ajustar “/path/directorio\_entrada” y
“/path/directorio\_salida” con los directorios reales en tu sistema.

3. Automatización de tareas con shell scripting

* Ejemplo: Crear un script para realizar una copia de seguridad de un
  directorio.




````bash
%%bash

# --- Configuración ---
# Detener el script si hay errores (-e) o variables no definidas (-u)
set -eu

FECHA=$(date +"%Y%m%d_%H%M%S")
ORIGEN="/content/sample_data"
# Definimos el destino dentro de /content para que lo veas en el panel izquierdo
DESTINO="/content/backup/backup_$FECHA"

# --- Validaciones ---

# 1. Verificar si el directorio de origen de Colab existe
if [ ! -d "$ORIGEN" ]; then
    echo "Error: El directorio de origen no existe en esta instancia: $ORIGEN" >&2
    exit 1
fi

# 2. Asegurar que el directorio contenedor del backup exista en /content
mkdir -p "/content/backup"

# --- Ejecución ---
echo "Iniciando copia de seguridad de los datos de muestra de Colab..."

# Copiar recursivamente (-r) y manteniendo atributos (-p)
if cp -r -p "$ORIGEN" "$DESTINO"; then
    echo "Copia de seguridad realizada correctamente."
    echo "Guardado en: $DESTINO"
    echo ""
    echo "--- Contenido respaldado ---"
    ls -lh "$DESTINO"
else
    echo "Ocurrió un error al copiar los archivos." >&2
    exit 1
fi
````



* Ejemplo: Comprimir todos los archivos de un directorio en un archivo
  tar.




````bash
%%bash

# --- Configuración ---
# Detener el script si hay errores (-e) o variables no definidas (-u)
set -eu

FECHA=$(date +"%Y%m%d_%H%M%S")
ORIGEN="/content/sample_data"
DIRECTORIO_DESTINO="/content/backup"
ARCHIVO_COMPRIMIDO="$DIRECTORIO_DESTINO/backup_$FECHA.tar.gz"

# --- Validaciones ---

# 1. Verificar si el directorio de origen de Colab existe y no está vacío
if [ ! -d "$ORIGEN" ] || [ -z "$(ls -A "$ORIGEN")" ]; then
    echo "Error: El directorio de origen no existe o está vacío: $ORIGEN" >&2
    exit 1
fi

# 2. Asegurar que el directorio de destino exista
mkdir -p "$DIRECTORIO_DESTINO"

# --- Ejecución ---
echo "Iniciando la compresión de los archivos de muestra..."

# Creamos el archivo tar.gz
# -c: crear, -z: comprimir con gzip, -v: verbose (listar archivos), -f: nombre del archivo
if tar -czvf "$ARCHIVO_COMPRIMIDO" -C "$ORIGEN" . ; then
    echo ""
    echo "Archivos comprimidos correctamente."
    echo "Ubicación del respaldo: $ARCHIVO_COMPRIMIDO"
    echo ""
    echo "--- Tamaño del archivo generado ---"
    ls -lh "$ARCHIVO_COMPRIMIDO"
else
    echo "Ocurrió un error durante la compresión." >&2
    exit 1
fi

````

* Ejemplos prácticos de scripting en Python para automatizar
  tareas: Ejemplo 1: Script de copia de seguridad utilizando
  Python:




````bash
import os
import shutil
import datetime

# --- Configuración de rutas ---
# Directorio de origen de Colab
directorio_origen = '/content/sample_data'

# Directorio de destino del respaldo
directorio_destino = '/content/backup'

# --- Validaciones de Seguridad ---
# 1. Verificar si el directorio de origen existe y tiene archivos
if not os.path.exists(directorio_origen):
    print(f"Error: El directorio de origen no existe: {directorio_origen}")
elif not os.listdir(directorio_origen):
    print(f"Advertencia: El directorio {directorio_origen} está vacío.")
else:
    try:
        # 2. Asegurar que el directorio de destino exista (reemplaza al mkdir -p)
        os.makedirs(directorio_destino, exist_ok=True)

        # 3. Generar la estampa de tiempo para el nombre (AñoMesDía_HoraMinutoSegundo)
        fecha_actual = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")

        # OJO: shutil.make_archive añade la extensión (.tar.gz) automáticamente.
        # Por lo tanto, el nombre base no debe llevar el '.tar.gz' al final.
        nombre_base_respaldo = os.path.join(directorio_destino, f'backup_{fecha_actual}')

        print("Iniciando la compresión de los archivos con Python...")

        # 4. Crear el archivo comprimido (.tar.gz)
        # 'gztar' equivale a un archivo tar comprimido con gzip
        archivo_creado = shutil.make_archive(
            base_name=nombre_base_respaldo,
            format='gztar',
            root_dir=directorio_origen
        )

        # --- Verificación del resultado ---
        print("\n Copia de seguridad realizada correctamente en Python.")
        print(f" Ubicación del respaldo: {archivo_creado}")

        # Mostrar el tamaño del archivo generado para control
        tamano_bytes = os.path.getsize(archivo_creado)
        tamano_mb = tamano_bytes / (1024 * 1024)
        print(f" Tamaño del archivo: {tamano_mb:.2f} MB")

    except Exception as e:
        print(f" Ocurrió un error inesperado durante el respaldo: {e}")
````



Este script utiliza las bibliotecas `shutil` y
`os` de Python para realizar una copia de seguridad de un
directorio especificado. El directorio se comprime en un archivo tar.gz
con el nombre del respaldo y la fecha actual.

Ejemplo 2: Script para procesar archivos CSV y generar un informe
utilizando Python:




````bash
import csv
import os

# --- Configuración de rutas en Colab ---
# Creamos el archivo de entrada dentro de la carpeta temporal de Colab
archivo_csv = '/content/usuarios_prueba.csv'
archivo_informe = '/content/informe_categorias.txt'


# --- PASO PREVIO: Crear un CSV de prueba en Colab ---
# (Esto simula el archivo que subirías o tendrías disponible)
datos_ejemplo = [
    ['Nombre', 'Edad'],
    ['Ruben', '25'],
    ['Ana', '17'],
    ['Carlos', '42'],
    ['Maria', '15']
]

print("Generando archivo CSV de prueba...")
with open(archivo_csv, 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(datos_ejemplo)


# --- PROCESAMIENTO DEL CSV Y GENERACIÓN DEL INFORME ---
print("Procesando datos y generando informe...")

if not os.path.exists(archivo_csv):
    print(f" Error: El archivo CSV no existe en la ruta: {archivo_csv}")
else:
    try:
        # Abrimos el CSV en modo lectura ('r') y el informe en modo escritura ('w')
        # Añadimos encoding='utf-8' para evitar problemas con tildes o caracteres especiales
        with open(archivo_csv, 'r', encoding='utf-8') as csv_file, \
             open(archivo_informe, 'w', encoding='utf-8') as informe_file:

            csv_reader = csv.DictReader(csv_file)

            # Escribir un encabezado bonito en el informe
            informe_file.write("=========================================\n")
            informe_file.write("      INFORME DE CATEGORÍAS POR EDAD      \n")
            informe_file.write("=========================================\n\n")

            contador_procesados = 0

            for row in csv_reader:
                # Extraemos y limpiamos los datos
                nombre = row['Nombre'].strip()
                edad = int(row['Edad'].strip())

                # Lógica de categorización
                categoria = 'Mayor de edad' if edad >= 18 else 'Menor de edad'

                # Escribir los resultados formateados en el archivo txt
                informe_file.write(f"• Nombre: {nombre:<10} | Categoría: {categoria} ({edad} años)\n")
                contador_procesados += 1

            informe_file.write(f"\nTotal de registros procesados: {contador_procesados}\n")

        # --- Verificación del Resultado ---
        print("\n Informe generado con éxito.")
        print(f" Ubicación del informe: {archivo_informe}")
        print("\n--- Contenido del informe generado: ---")

        # Mostrar el contenido del txt directamente en la consola de Colab
        with open(archivo_informe, 'r', encoding='utf-8') as f:
            print(f.read())

    except KeyError as ke:
        print(f" Error de formato: No se encontró la columna {ke} en el archivo CSV.")
    except ValueError:
        print(" Error de datos: Asegúrate de que la columna 'Edad' contenga solo números enteros.")
    except Exception as e:
        print(f" Ocurrió un error inesperado: {e}")
````



Este script utiliza la biblioteca `csv` de Python para
procesar un archivo CSV y generar un informe con ciertos cálculos o
manipulaciones de datos. En este ejemplo, se clasifica a las personas
como “Mayor de edad” o “Menor de edad” según su edad.

---

* Presentación de Ansible como herramienta de automatización y

  utiliza para automatizar la configuración, implementación y
  administración de sistemas. Permite describir la infraestructura como

* Explicación de los conceptos clave de Ansible:

  + Inventario: Es un archivo que contiene la lista de los hosts en los
    que se realizarán las tareas. Puede ser un archivo de texto o una base
    de datos dinámica.
  + Playbooks: Son archivos escritos en formato YAML que describen las
    tareas y configuraciones a realizar en los hosts. Contienen una serie de
    tareas y se ejecutan en orden.
  + Módulos: Son piezas de código que Ansible utiliza para realizar
    tareas en los hosts. Ansible proporciona una amplia gama de módulos que
    cubren diversas funcionalidades, como gestión de paquetes, creación de
    usuarios, configuración de servicios, entre otros.
  + Roles: Son unidades organizativas que agrupan tareas relacionadas en
    un playbook. Permiten estructurar y reutilizar el código de manera
    modular.
* Ejemplos prácticos de uso de Ansible para la automatización de
  tareas en Unix/Linux: Ejemplo 1: Instalación de paquetes en hosts
  remotos utilizando Ansible:




```
---
- name: Instalar paquetes
  hosts: webservers
  tasks:
    - name: Actualizar repositorios
      apt:
        update_cache: yes

    - name: Instalar paquetes
      apt:
        name:
          - apache2
          - php
        state: present

```




Este playbook de Ansible realiza la instalación de los paquetes
Apache2 y PHP en los hosts definidos en el inventario. Utiliza el módulo
`apt` para gestionar los paquetes en sistemas basados en
Debian/Ubuntu.

Ejemplo 2: Configuración de un servidor Nginx utilizando Ansible:




```
---
- name: Configurar servidor Nginx
  hosts: webservers
  tasks:
    - name: Instalar Nginx
      apt:
        name: nginx
        state: present

    - name: Copiar archivo de configuración
      copy:
        src: files/nginx.conf
        dest: /etc/nginx/nginx.conf
        owner: root
        group: root
        mode: 0644

    - name: Reiniciar servicio Nginx
      service:
        name: nginx
        state: restarted

```




Este playbook de Ansible instala Nginx en los hosts definidos en el
inventario, copia un archivo de configuración personalizado y reinicia
el servicio Nginx para aplicar los cambios.

* Comparativa entre scripts tradicionales y Ansible en términos de
  funcionalidad y mantenimiento:
  + Funcionalidad: Ansible proporciona una estructura y una forma más
    declarativa de definir y ejecutar tareas, lo que facilita la
    automatización y la reutilización de código. Además, cuenta con una
    amplia gama de módulos que cubren muchas funcionalidades comunes en
    sistemas Unix/Linux.
  + Mantenimiento: Los playbooks de Ansible son fáciles de leer,
    entender y mantener, ya que están escritos en formato YAML y siguen una
    estructura clara. Además, Ansible gestiona el estado de los sistemas y
    solo aplica los cambios necesarios, lo que facilita la administración y
    el mantenimiento a largo plazo.

Estos ejemplos son solo una introducción básica a Ansible. La
herramienta ofrece muchas más funcionalidades y características
avanzadas que pueden ser exploradas para la automatización de tareas en
entornos Unix/Linux.

Para ejecutar los ejemplos de Ansible, sigue los siguientes
pasos:

1. Instala Ansible: Asegúrate de tener Ansible instalado en tu
   sistema. Puedes seguir la documentación oficial de Ansible para
   instalarlo según tu sistema operativo: <https://docs.ansible.com/ansible/latest/installation_guide/index.html>
2. Crea un inventario: Crea un archivo de inventario que liste los
   hosts en los que deseas ejecutar las tareas. Puedes utilizar un archivo
   de texto simple o una base de datos dinámica según tus necesidades. Por
   ejemplo, crea un archivo llamado `inventory.ini` con el
   siguiente contenido:




```
[webservers]
host1 ansible_host=192.168.0.1
host2 ansible_host=192.168.0.2

```




Asegúrate de reemplazar las direcciones IP con las direcciones reales
de los hosts.

3. Crea el playbook: Crea un archivo YAML que contenga el playbook
   de Ansible con las tareas que deseas ejecutar en los hosts. Puedes
   utilizar los ejemplos proporcionados o crear tus propios playbooks.
   Guarda el archivo con la extensión `.yml`. Por ejemplo,
   guarda el primer ejemplo como `install_packages.yml` y el
   segundo ejemplo como `configure_nginx.yml`.
4. Ejecuta el playbook: Abre una terminal y navega hasta el
   directorio donde tienes el archivo de inventario y el playbook. Luego,
   ejecuta el siguiente comando para ejecutar el playbook:




```
ansible-playbook -i inventory.ini install_packages.yml

```




Asegúrate de reemplazar `inventory.ini` con el nombre real
de tu archivo de inventario y `install_packages.yml` con el
nombre real de tu playbook. Esto ejecutará el playbook en los hosts
especificados en el inventario y realizará las tareas
correspondientes.

Puedes ejecutar otros playbooks de la misma manera, reemplazando el
nombre del playbook en el comando.

Estos son solo pasos básicos para ejecutar los ejemplos. Ansible
ofrece muchas más funcionalidades y opciones avanzadas que puedes
explorar para adaptar la ejecución a tus necesidades específicas.
